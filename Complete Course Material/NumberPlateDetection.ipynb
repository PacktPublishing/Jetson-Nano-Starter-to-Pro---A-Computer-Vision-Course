# **Training on Custom Dataset**

### **VOC Dataset Formetting**

```python
import os
import random

trainval_percent = 0.1
train_percent = 0.9
xmlfilepath = '/content/drive/MyDrive/YOLOX/Our Dataset/Labels'
txtsavepath = '/content/drive/MyDrive/YOLOX/Our Dataset/Images'
total_xml = os.listdir(xmlfilepath)

num = len(total_xml)
list = range(num)
tv = int(num * trainval_percent)
tr = int(tv * train_percent)
trainval = random.sample(list, tv)
train = random.sample(trainval, tr)

ftest = open('/content/drive/MyDrive/YOLOX/Our Dataset/Text/tes.txt', 'w')
ftrain = open('/content/drive/MyDrive/YOLOX/Our Dataset/Text/trainval.txt', 'w')

for i in list:
    name = total_xml[i][:-4] + '\n'
    if i in trainval:
        ftest.write(name)
    else:
        ftrain.write(name)

ftrain.close()
ftest.close()
```
